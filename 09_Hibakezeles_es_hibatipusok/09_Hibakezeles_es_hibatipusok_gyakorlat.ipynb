#  1. Feladat megoldás

```python
betu = input("Adj meg egy betűt: ")

try:
  szam = int(betu)
except ValueError:
  print("A karaktert nem lehet egész számmá átlakítani!")
```

# 2. Feladat megoldás

```python
betu = input("Adj meg egy betűt: ")

try:
  eremeny = betu + 0
except TypeError:
  print("A karakterhez nem lehet egész számot hozzáadni!")
```

# 3. Feladat megoldás

```python
szam = float(input("Adj meg egy számot: "))

try:
  eremeny = szam / 0
except ZeroDivisionError:
  print("A bekért számot nem lehet nullával elosztani!")
```

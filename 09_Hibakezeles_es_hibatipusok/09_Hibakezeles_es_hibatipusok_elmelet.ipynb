# Példák

## Lista értékeinek reciproka

````python
eredeti_lista = [1, 5, 2, 0, 10, 20]
reciprok_lista = []

for elem in eredeti_lista:
  reciprok_lista.append(1 / elem)

print(reciprok_lista)
````

````python
eredeti_lista = [1, 5, 2, 0, 10, 20]
reciprok_lista = []

try:
  for elem in eredeti_lista:
    reciprok_lista.append(1 / elem)
except ZeroDivisionError as e:
  print(f'Hiba: {e}')

print(reciprok_lista)
````

````python
eredeti_lista = [1, 5, 2, 0, 10, 20]
reciprok_lista = []

for elem in eredeti_lista:
  try:
    reciprok_lista.append(1 / elem)
  except ZeroDivisionError as e:
    print(f'Hiba: {e}')

print(reciprok_lista)
````

## Felasználótól bekért érték vizsgálata

````python
szam = int(input('Adjon meg egy egész számot: '))
````

````python
try:
  szam = int(input('Adj meg egy egész számot: '))
except ValueError as e:
  print(f'Nem megfelelő bemenet.')
````

````python
szam = False

while not szam:
  try:
    szam = int(input('Adj meg egy egész számot: '))
  except ValueError as e:
    print(f'Nem megfelelő bemenet.')

print(f'Megadott szám: {szam}')
````

## Több except blokk

````python
eremeny = None

try:
  szam = int(input('Adj meg egy egész számot: '))
  eremeny = 10 / szam
except ValueError as e:
  print(f'Nem megfelelő bemenet.')
except ZeroDivisionError as e:
  print(f'Osztás 0-val nem megengedett.')

print(f'Eredmény: {eremeny}')
````

# 1. Feladat

- Hozz létre egy üres listát
- Kérj be 2 szót és add hozzá őket a listához
- Rendezd a listát, majd írd ki az eredményt

Példa:
> Egyik szó: barack  
> Másik szó: alma  
> Rendezett lista: ['alma', 'barack']

# 2. Feladat

- Hozz létre egy üres listát
- Kérj be 3 számot és mentsd el őket a listába
- Számold ki az elemek összegét, majd add hozzá a listához
- Írd ki az így kapott új listát

Példa:
> Első elem: 4  
> Második elem: 6  
> Hardmadik elem: -1  
> Lista kiegészítve az összeggel: [4, 6, -1, 9]

# 3. Feladat

- Hozz létre egy üres listát
- Kérj be 4 számot és mentsd el őket a listába
- Töröld ki a duplikált elemeket
- Írd ki a lista elemeit csökkenő sorrendben

Példa:
> Első elem: 4  
> Második elem: 6  
> Hardmadik elem: 6  
> Negyedik elem: -1  
> Duplikált nélkül, csökkenő sorrendben: [6, 4, -1]

# 4. Feladat

- Kérj be két nevet és egy-egy hozzájuk tartozó kedvenc italt
- Mentsd el ezeket egy két elemű dictionary-be (ahol a név a kulcs, a kedvenc ital pedig az érték)
- Kérj be egy korábbi nevet és add meg mi volt a kedvenc itala

Példa:
> Első név: Sámuel  
> Kedvenc itala: Pepsi  
> Második név: Mózes  
> Kedvenc itala: Víz  
> Vizsgált személy: Sámuel  
> Sámuel kedvenc itala: Pepsi

# 1. Feladat megoldás

```python
# Üres lista létrehozása
lista = []

# Adatok bekérése és listához adása
egyik_szo = input('Egyik szó: ')
lista.append(egyik_szo)

masik_szo = input('Másik szó: ')
lista.append(masik_szo)

# Elemek rendezése és kiírás
rendezett_lista = sorted(lista)
print('Rendezett lista: ', rendezett_lista)
```

# 2. Feladat megoldás

```python
# Üres lista létrehozása
lista = []

# Adatok bekérése és listához adása
elso_elem = int(input('Első elem: '))
lista.append(elso_elem)

masodik_elem = int(input('Második elem: '))
lista.append(masodik_elem)
             
harmadik_elem = int(input('Hardmadik elem: '))
lista.append(harmadik_elem)


# Összeg hozzáadása és kiírás
lista.append(sum(lista))

print('Lista kiegészítve az összeggel: ', lista)
```

# 3. Feladat megoldás

```python
# Üres lista létrehozása
lista = []

# Adatok bekérése és listához adása
elso_elem = int(input('Első elem: '))
lista.append(elso_elem)

masodik_elem = int(input('Második elem: '))
lista.append(masodik_elem)
             
harmadik_elem = int(input('Hardmadik elem: '))
lista.append(harmadik_elem)

negyedik_elem = int(input('Negyedik elem: '))
lista.append(negyedik_elem)

# Duplikálak eltávoltása, csökkenő sorrend és kiírás
lista = list(set(lista))
lista.sort()
lista.reverse()

print('Duplikált nélkül, csökkenő sorrendben: ', lista)
```

# 4. Feladat megoldás

```python
# Adatok bekérése és mentése
elso_nev = input("Első név: ")
elso_kedvenc_ital = input("Kedvenc itala: ")
masodik_nev = input("Második név: ")
masodik_kedvenc_ital = input("Kedvenc itala: ")

kedvenc_italok = {elso_nev: elso_kedvenc_ital, masodik_nev: masodik_kedvenc_ital}

# Vizsgált személy bekérése és eremény kiírása
vizsglat_szemely = input("Vizsgált személy: ")
print(f"{vizsglat_szemely} kedvenc itala: {kedvenc_italok[vizsglat_szemely]}")
```

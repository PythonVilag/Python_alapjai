## Fontosabb f√ºggv√©nyek

Fontos vigy√°zni, mert vannak 'in-place' f√ºggv√©nyek, amiket lefuttatva megv√°ltozik az eredeti v√°ltoz√≥nk √©rt√©ke, de vannak olyanok is amiket egy √∫j v√°ltoz√≥ba el kell menteni (vagy az eredetit fel√ºl√≠rni), hogy hat√°suk legyen.

N√©h√°ny a n√©pszer≈±bb list√°kon futtathat√≥ f√ºggv√©nyek k√∂z√ºl:


- `<lista>.insert(<index>, <elem>)` ‚Üí besz√∫r egy elemet a megadott index hely√©re (in-place)

- `<lista>.remove(<elem>)` ‚Üí elt√°vol√≠tja az elemet a list√°b√≥l (in-place)

- `<lista>.reverse()` ‚Üí megford√≠tja az elemek sorrendj√©t (in-place)

- `<lista>.count(<keresett elem>)` ‚Üí megadja, hogy keresett elem h√°nyszor van jelen a list√°ban

- `<lista>.index(<keresett elem>)` ‚Üí megadja, hogy keresett elem a lista h√°nyadik eleme

- `<sz√∂veg>.split(<elv√°laszt√≥ sz√∂veg>)` ‚Üí a k√ºl√∂n v√°lasztott szavakat egy lista egym√°st k√∂vet≈ë elemeibe pakolja. Ha nincs megadva elv√°laszt√≥ sz√∂veg akkor white-spacenk√©nt (space, tab, enter) v√°lt elemet

- `<sz√∂veg>.join(list(<egyik sz√∂veg>, <m√°sik sz√∂veg>, <harmadik sz√∂veg>))` ‚Üí a list√°ban jelenl√©v≈ë sz√∂vegeket egym√°s mell√© f≈±zni a sz√∂veg stringgel elv√°lasztva

- `<elem> in <lista>` ‚Üí megadja, hogy az adott elem megtal√°lhat√≥ e valahol a list√°ban (üêü)

````python
lista.insert(2, '√∫j kettes index≈± elem')

print(lista)
````

````python
lista.remove('√∫j elem')

print(lista)
````

# Dictionary

- A dictionary l√©nyeg√©ben ugyanaz, mint a lista, azzal a k√ºl√∂nbs√©ggel, hogy itt ez elemekre a sorsz√°muk helyett kulcs √©rt√©kekkel hivatkozunk.

- Ez azt is jelenti, hogy deklar√°l√°skor meg kell adni a kulcsokat √©s a hozz√°juk tartoz√≥ √©rt√©keket is.

## L√©trehoz√°s √©s tulajdons√°gok

- L√©trehoz√°skor az elemp√°rokat kapcsos z√°r√≥jelek `{}` k√∂z√© tessz√ºk

- Egy elemp√°r egy kulcsb√≥l √©s egy ahhoz tartoz√≥ √©rt√©kb≈ël √°ll

- Az elemp√°r elemeit kett≈ësponttal v√°lasztjuk el egym√°st√≥l: `<kulcs>: <√©rt√©k>`

- Az elemp√°rokat vessz≈ëvel v√°lasztjuk el egym√°st√≥l

- Az √©rt√©kekre sorsz√°muk helyett a kulcs seg√≠ts√©g√©vel hivatkozunk

- A dictionary elemeit meg tudjuk v√°ltoztatni

```python
<dictionary neve> = {<els≈ë kulcs>: <els≈ë √©rt√©k>, ..., <utols√≥ kulcs>: <utols√≥ √©rt√©k>}
```

````python
barat = {'n√©v': 'M√°rton', 'kor': 34, 'hobby': [], 'vicces': True, 'kedvenc s√∂r': ['Borsodi', 'Soproni']}
````

L√©trehozhatn√°nk egy list√°t is, ugyanezekkel az √©rt√©kekkel, csak ott index seg√≠ts√©g√©vel tudn√°nk r√°juk hivatkozni kulcs helyett:
```python
barat = ['M√°rton', 34, [], True, ['Borsodi', 'Soproni']]
```

## M≈±veletek

**Hivatkoz√°s elemre kulcs haszn√°lat√°val**


````python
barat['n√©v']
````

**Hivatkoz√°s elemre index haszn√°lat√°val (nem lehet!)**

```python
barat[0]
```

<sup>Persze ha l√©trehozunk egy kulcsot egy index √©rt√©k√©vel akkor lehet, de val√≥j√°ban akkor is a kulccsal hivatkozunk az elemre.</sup>

**Elemek friss√≠t√©se, b≈ëv√≠t√©se, t√∂rl√©se (üêü)**

````python
barat['kor'] = 25

print(barat)
````

````python
del(barat['hobby'])

print(barat)
````

## Fontosabb f√ºggv√©nyek (üêü)

N√©h√°ny a n√©pszer≈±bb dictionary-ken futtathat√≥ f√ºggv√©nyek k√∂z√ºl:

- `len(<dictionary neve>)` ‚Üí megadja a dictionary-ban tal√°lhat√≥ kulcsok sz√°m√°t

- `<dictionary neve>.get(<kulcs>)` ‚Üí megadja a kulcshoz tartoz√≥ √©rt√©ket, de nem dob hib√°t ha a kulcs nem tal√°lhat√≥

- `<dictionary neve>.keys()` ‚Üí visszaadja a dictionary-ben t√°rolt kulcsokat

- `<dictionary neve>.values()` ‚Üí visszaadja a dictionary-ben t√°rolt √©rt√©keket

- `<dictionary neve>.items()` ‚Üí visszaadja a dictionary-ben t√°rolt kulcs-√©rt√©k p√°rokat

- `<keresett kulcs> in <dictionary neve>` ‚Üí megadja, hogy az adott kulcs megtal√°lhat√≥ e valahol a dictionary-ben

````python
len(barat)
````

````python
print(barat.get('magass√°g'))
````

````python
barat.keys()
````

````python
barat.values()
````

````python
barat.items()
````

````python
'vicces' in barat
````

## Set

### L√©trehoz√°s √©s tulajdons√°gok

- L√©trehoz√°skor az elemeket kapcsos z√°r√≥jelek `{}` k√∂z√© tessz√ºk


````python
szamok = {0, 1, 2, 2, 4, 6, 6}
````

````python
szamok
````

# 1. Feladat

- Kérj be egy szót.
- Írd ki a bekért szót kétszer
- Adj hozzá legalább 2 kommentet a kódhoz, ahol elmagyarázod mi történik

Példa:
> Szöveg: Yo  
> Duplázott szöveg: YoYo

# 2. Feladat

- Kérj be egy születési évet.
- Írd ki a felhasználó nevét és hogy hány éves lesz 2030-ban

> Felhasználó neve: János  
> Születési év: 2001  
> János 2030-ban 29 éves lesz.

# 3. Feladat

- Kérj be két tört számot
- Írd ki a két szám szorzatát egy f-stringben 2 tizedes pontossággal
- Adj hozzá legalább 3 kommentet a kódhoz, ahol elmagyarázod mi történik

Példa:
> Egyik szám: 11.3  
> Másik szám: 2.11  
> A két bekért szám szorzata: 23.84

# 1. Feladat megoldás

```python
# Adatok bekérése
szoveg = input('Szöveg: ')

# Eredmény számítása és kiírása
szoveg *= 2
print(f'Duplázott szöveg: {szoveg}')
```

# 2. Feladat megoldás

```python
# Adatok bekérése
nev = input('Felhasználó neve: ')
szuletei_ev = int(input('Születési év: '))

# Eredmény kiírása
print(f'{nev} 2030-ban {2030 - szuletei_ev} éves lesz.')
```

# 3. Feldat megoldás

```python
# Adatok bekérése
szam_1 = input('Egyik szám: ')
szam_2 = input('Másik szám: ')

# Bekért adatok valós számmá alakítása
szam_1 = float(szam_1)
szam_2 = float(szam_2)

# Eredmény kiírása 2 tízedes pontossággal
print(f'A két bekért szám szorzata: {szam_1 * szam_2:0.2f}')
```

# 1. Feladat

- Kérj be két valós számú értéket
- Írasd ki a két szám szorzatát egy f-stringben 2 tízedes pontossággal
- Adj hozzá legalább 3 kommentet a kódhoz, ahol elmagyarázod mi történik

Példa:
> Egyik szám: 11.3  
> Másik szám: 2.11  
> A két bekért szám szoraza: 23.73

# 1. Feladat megoldás

```python
# Adatok bekérése
szam_1 = input('Egyik szám: ')
szam_2 = input('Másik szám: ')

# Bekért adatok valós számmá alakítása
szam_1 = float(szam_1)
szam_2 = float(szam_2)

# Adatok kiírása 2 tízedes pontossággal
print(f'A két bekért szám szoraza: {szam_1 * szam_2:.2f}')
```

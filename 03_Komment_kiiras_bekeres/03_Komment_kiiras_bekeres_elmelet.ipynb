# Kiírás

- Adatok közlése
- Program működésének ellenőrzése (debuggolás)
- A függvénynek zárójelek között adjuk meg, mit írjon ki
- Ha több elemet is ki szeretnénk iratni vesszővel elválasztva tehetjük ezt meg (ezek lehetnek különböző típusúak is)
- Több elem esetén a print függvény space-el választja el az elemetek (alapértelmezetten)

````python
print('Hello, világ!')
````

````python
print('Gyula életkora ' + '23' + '!')    # Csak 1 elem, mivel a print csak 1 stringet lát (ami a többinek az összefűzése)
````

````python
# Szöveg és szám kiírása
print('Gyula életkora', 23, '!')    # 3 elem, vesszővel elválasztva
````

<sup>Itt megfigyelhető az extra space a 23 és a felkiáltójel között</sup>

````python
print('Gyula életkora ' + str(23) + '!')    # Típus konverzió után már összefűzhetőek a string típusú adatok
````

Kiírásra a későbbiekben 'formatted string'-eket, úgynevezett f-stringeket fogunk használni. Az f-string olyan szöveg, amely típus konverziók nélkül a szövegbe intergrálva tud kiírni nem szöveg típusú változót, vagy akár műveleti eredményt is.

Használatát a kezdő aposztróf elé tett *f* karakterrel jelöljük, és az integrálandó adatot kapcsos zárójelek *{ }* közé tesszük.

````python
kor = 25
most = 2022
````

````python
f_string = f'Dani idén tölti {kor}. életévét, mert {most - kor}-ben született.'

print(f_string)
````

f-stringek esetén tört számoknál a kerekítés pontosságát is meg lehet adni.

````python
szorzat = 13.334*17.951
print(szorzat)
````

````python
print(f'Szorzat 3 tizedes pontossággal: {szorzat:.3f}')
````

Numerikussá alakítva a bemenetet

````python
szam_1 = input('Szám 1: ')
szam_2 = input('Szám 2: ')

# Számok átalakítása numerikus értékké
szam_1 = float(szam_1)
szam_2 = float(szam_2)
````

````python
osszeg = szam_1 + szam_2

print(osszeg)
````

# Ki√≠r√°s

- Adatok k√∂zl√©se

- Program m≈±k√∂d√©s√©nek ellen≈ërz√©se (debuggol√°s)

## L√©trehoz√°s √©s tulajdons√°gok

- Ki√≠rni az be√©p√≠tett `print()` f√ºggv√©ny seg√≠ts√©g√©vel tudunk

- A f√ºggv√©nynek z√°r√≥jelek k√∂z√∂tt adjuk meg, mit √≠rjon ki

- Ha t√∂bb elemet is ki szeretn√©nk iratni vessz≈ëvel elv√°lasztva tehetj√ºk ezt meg (ezek lehetnek k√ºl√∂nb√∂z≈ë t√≠pus√∫ak is)

```python
print(<egyik ki√≠rand√≥ √©rt√©k>, <m√°sik ki√≠rand√≥ √©rt√©k>, ...)
```

**Sz√∂vek ki√≠r√°sa**

````python
print('Hello, vil√°g!')
````

````python
print('Gyula √©letkora' + '23' + '!')
````

**Sz√∂veg √©s sz√°m ki√≠r√°sa k√ºl√∂n**

````python
print('Gyula √©letkora', 23, '!')  # 3 elem, vessz≈ëvel elv√°lasztva
````

**Sz√∂veg √©s sz√°m ki√≠r√°sa konverzi√≥val**

````python
print('Gyula √©letkora' + str(23) + '!')  # T√≠pus konverzi√≥ ut√°n m√°r √∂sszef≈±zhet≈ëek a string t√≠pus√∫ adatok
````

## Az f-string

- Ki√≠r√°sra a k√©s≈ëbbiekben 'formatted string'-eket, √∫gynevezett f-stringeket fogunk haszn√°lni. 

- Az f-string olyan sz√∂veg, amely t√≠pus konverzi√≥k n√©lk√ºl a sz√∂vegbe integr√°lva tud ki√≠rni v√°ltoz√≥t, vagy ak√°r m≈±veleti eredm√©nyt is.

- Haszn√°lat√°t a kezd≈ë aposztr√≥f el√© tett `f` karakterrel jel√∂lj√ºk, √©s az integr√°land√≥ adatot kapcsos z√°r√≥jelek `{ }` k√∂z√© tessz√ºk.

````python
kor = 46
most = 2022
````

````python
f_string = f'Laci id√©n t√∂lti {kor}. √©let√©v√©t, mert {most - kor}-ben sz√ºletett.'

print(f_string)
````

f-stringek eset√©n t√∂rt sz√°mokn√°l a kerek√≠t√©s pontoss√°g√°t is meg lehet adni.

````python
szorzat = 13.334 * 17.951
print(szorzat)
````

````python
print(f'Szorzat 3 tizedes pontoss√°ggal: {szorzat:.3f}')
````

## Print f√ºggv√©ny param√©terek (üêü)

Vizsg√°ljuk meg a ki√≠r√°st az al√°bbi szempontokb√≥l:
- Ki√≠rand√≥ √©rt√©kek elv√°laszt√°sa
- Ki√≠r√°s ut√°na karakter

````python
print("Jakabnak van", 3, "mogyor√≥ja.")
print("Nekem csak", 2, "mogyor√≥m van.")
````

Elv√°laszt√≥ karakter:     space `' '`  
Ki√≠r√°s ut√°ni karakter:   enter `'\n'`

Ezeket a param√©tereket m√≥dos√≠thatjuk a `sep` √©s az `end` f√ºggv√©ny param√©terek seg√≠ts√©g√©vel.

````python
print("Jakabnak van", 3, "mogyor√≥ja.", sep='?', end='\t')
print("Nekem csak", 2, "mogyor√≥m van.", sep='\n')
````

# Bek√©r√©s

- A felhaszn√°l√≥ fut√°s k√∂zben tud √©rt√©ket defini√°lni

## L√©trehoz√°s √©s tulajdons√°gok

- √ârt√©ket bek√©rni az `input()` f√ºggv√©ny seg√≠ts√©g√©vel lehet

- A f√ºggv√©nynek z√°r√≥jelek k√∂z√∂tt adjuk meg, hogy mit √≠rjon ki, mint bek√©r√©st seg√≠t≈ë utas√≠t√°s

- A bek√©rt √©rt√©ket el kell menteni egy v√°ltoz√≥ba, hogy k√©s≈ëbb haszn√°lni tudjuk

- A bek√©rt √©rt√©k mindig `str` t√≠pus√∫, ez√©rt ha m√°s form√°tum√∫ adatot szeretn√©nk haszn√°lni t√≠pus konverzi√≥t kell alkalmazni


```python
<v√°ltoz√≥ neve> = input(<bek√©r√©shez tartoz√≥ utas√≠t√°s>)
```

````python
nev = input('Mi a neved? ')
````

````python
print(f'Szia, ' + nev + ' !')
````

````python
kor = input('H√°ny √©ves vagy? ')
````

````python
print('A te korod: ' + kor)
````

# P√©ld√°k

## Nem m≈±k√∂d≈ë sz√°m √∂sszead√≥

````python
# Adatok bek√©r√©se
szam_1 = input('Sz√°m 1: ')
szam_2 = input('Sz√°m 2: ')
````

````python
# Eredm√©ny sz√°m√≠t√°sa √©s ki√≠r√°sa
osszeg = szam_1 + szam_2

print(szam_1 + ' + ' + szam_2 + ' = ' + osszeg)
````

## M≈±k√∂d≈ë sz√°m √∂sszead√≥

````python
# Adatok bek√©r√©se
szam_1 = input('Sz√°m 1: ')
szam_2 = input('Sz√°m 2: ')

# Bek√©rt adatok t√∂rt sz√°mm√° alak√≠t√°sa
szam_1 = float(szam_1)
szam_2 = float(szam_2)
````

````python
# Eredm√©ny sz√°m√≠t√°sa
osszeg = szam_1 + szam_2
````

```python
print(szam_1 + ' + ' + szam_2 + ' = ' + osszeg)
```

````python
# Eredm√©ny ki√≠r√°sa
print(f'{szam_1} + {szam_2} = {osszeg}')
````

<sup>A kimenet az√©rt 7.0 √©s nem 7, mert float t√≠pus√∫ sz√°mokat adunk √∂ssze nem int-eket.</sup>

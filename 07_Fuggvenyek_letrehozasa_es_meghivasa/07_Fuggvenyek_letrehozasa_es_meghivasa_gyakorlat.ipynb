# 4. Feladat

- Írj egy függvényt ami bekér 2 pozitív egész számot
- A függvény kiszámítja, hogy a két bekért szám relatív prím-e
- Visszaadja az eredményt bool formájában
- Írd ki a függvény kimenetét két felhasználótól bekért számra

Példa:
> Egyik szám: 24  
> Másik szám: 25  
> A 24 és 25 relatív prímek.

#  1. Feladat megoldás

```python
def harom_hatvanyai():
  for kitevo in range(4):
    print(f'3 ** {kitevo} = {3 ** kitevo}')

harom_hatvanyai()
```

#  2. Feladat megoldás

```python
def szam_hatvanyai(szam=3, hatvanyszam=4):
  for kitevo in range(hatvanyszam):
    print(f'{szam} ** {kitevo} = {szam ** kitevo}')

szam_hatvanyai(hatvanyszam=2)
```

# 3. Feladat megoldás

```python
def honapbol_evszak(honap_szam):
  evszakok = ['tél', 'tavasz', 'nyár', 'ősz']
  evszak_szam = (honap_szam % 12) // 3

  return evszakok[evszak_szam]

honap_szam = int(input('Évszak meghatározása az alábbi hónapra: '))

print(f'A {honap_szam}. hónaphoz tartozó évszak: {honapbol_evszak(honap_szam)}.')
```

# 4. Feladat megoldás

```python
def relativ_primek(szam_1, szam_2):
  oszto_szam = 0
  for oszto in range(1, min(szam_1, szam_2) + 1):
    if (szam_1 % oszto == 0) and (szam_2 % oszto == 0):
      oszto_szam += 1
  if oszto_szam == 1:
    return True
  else:
    return False

szam_1 = int(input('Egyik szám: '))
szam_2 = int(input('Másik szám: '))

if relativ_primek(szam_1, szam_2):
  print(f'A {szam_1} és {szam_2} relatív prímek.')
else:
  print(f'A {szam_1} és {szam_2} nem relatív prímek.')
```

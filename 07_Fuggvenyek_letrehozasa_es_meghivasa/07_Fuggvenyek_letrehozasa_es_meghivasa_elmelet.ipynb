# Függvények

- Egy adott feladat elvégzését tartalmazó kódrészlet
- Csak 1x kell megírni a kódot és több helyen lehet használni

Struktúra:
```python
def fuggveny_neve(bemenet_1, bemenet_2, ...):
  számold ki a visszatérési értéket a bemenetek segítségével
  return visszatérési érték
```

Már korábban is alkalmaztunk függvényeket:
- input()
- print()
- len()

Függvény, ami hatványoz

````python
def hatvany_szamitas(alap, kitevo):
  eredmeny = alap**kitevo
  return eredmeny
````

````python
eredmeny = hatvany_szamitas(3, 2)
````

````python
eredmeny
````

````python
hatvany_szamitas(kitevo=3, alap=4)
````

````python
def osztok_szama(szam):
  oszto_darabszam = 0
  for oszto in range(1, szam+1):
    if szam % oszto == 0:
      oszto_darabszam += 1
  return oszto_darabszam
````

````python
osztok_szama(3)
````

# Elágazás (if)

- Lehetőséget biztosít, hogy különböző feltételek mellett a programunk máshogyan viselkedjen

Struktúra:
```python
if <egyik feltétel>:
  <csinálj valamit ha "egyik feltétel" logikai érétke igaz>
elif <másik feltétel>:
  <csinálj valamit ha "egyik feltétel" hamis és "másik feltétel" igaz>
...
else:
  <csinálj valamit ha egyik korábbi feltétel sem igaz>
```

````python
if False:
  print('A feltétel nem igaz')
elif True:
  print('A feltétel igaz')
````

# El√°gaz√°s

- Lehet≈ës√©get biztos√≠t, hogy k√ºl√∂nb√∂z≈ë felt√©telek mellett a programunk m√°shogyan viselkedjen

- T√∂bb felt√©telt is szabhatunk, illetve azt is megadhatjuk mit csin√°ljon a program ha egyik felt√©tel sem teljes√ºl

## L√©trehoz√°s √©s tulajdons√°gok

- 3 kulcssz√≥ √°ll a rendelkez√©s√ºnkre el√°gaz√°sok kezel√©s√©re: `if`, `elif` √©s `else`

- Ezek k√∂z√ºl az utols√≥ kett≈ë opcion√°lis

- `if` √©s `elif` eset√©n egy felt√©telnek kell k√∂vetnie a kulcssz√≥t, amely ha igaz, akkor a felt√©telhez tartoz√≥ k√≥dr√©szlet ki√©rt√©kel√©sre ker√ºl

- A parancssorok v√©g√©n kett≈ëspont karakter (`:`) tal√°lhat√≥, √©s a k√∂vetkez≈ë sorok bentebb vannak h√∫zva

- A beh√∫z√°s jel√∂li, hogy mi az a k√≥dr√©szlet, ami akkor fut le, ha az el≈ëtte l√©v≈ë felt√©tel igaz volt

- A parancshoz tartoz√≥ alk√≥d a k√©s≈ëbbiekben is t√∂bbsz√∂r el≈ë fog m√©g j√∂nni

```python
if <egyik felt√©tel>:
  <csin√°lj valamit ha <egyik felt√©tel> logikai √©rt√©ke igaz>
elif <m√°sik felt√©tel>:
  <csin√°lj valamit ha <egyik felt√©tel> hamis √©s <m√°sik felt√©tel> igaz>
...
else:
  <csin√°lj valamit ha egyik kor√°bbi felt√©tel sem igaz>
```

<sup>A felt√©tel ut√°n kett≈ëspont van √©s az adott felt√©telhez tartoz√≥ k√≥d bentebb van h√∫zva!</sup>

<sup>Ez jelenti azt, hogy melyik az a sor, amelyik m√©g az el√°gaz√°shoz tartozik √©s melyik, ami m√°r nem.</sup>

````python
jo_a_kedvem = True

if jo_a_kedvem:
  print('√út√∂m a kezemet :)')
else:
  print('Nem √ºt√∂m a kezemet :(')

print('Ez a sor a felt√©telekt≈ël f√ºggetlen√ºl lefut.')
````

## Program fut√°s√°nak meger≈ës√≠t√©se

````python
megerosites = input('Szeretn√©d, hogy a k√≥d lefusson? (igen/nem): ')

if megerosites == 'igen':
  print('Program fut. Tal√°lkozunk 30 perc m√∫lva!')

if megerosites == 'nem':
  print('A k√≥d nem fut le.')

else:
  print('Nem megfelel≈ë bemeneti √©rt√©k!')
````

Fontos, hogy az `else` mindig csak a legut√≥bbi `if` - `elif` sorozatot veszi figyelembe. Vagyis itt elindult egy "sorozat" az `if megerosites == 'igen'` sorral, majd egyb≈ël kezd≈ëd√∂tt egy √∫jabb az `if megerosites == 'nem'`-el.

Ha az eg√©szet egy sorozatk√©nt akarjuk defini√°lni, akkor a m√°sodik `if`-et `elif`-re kell cser√©ln√ºnk.

````python
megerosites = input('Szeretn√©d, hogy a k√≥d lefusson? (igen/nem): ')

if megerosites == 'igen':
  print('Program fut. Tal√°lkozunk 30 perc m√∫lva!')
  
elif megerosites == 'nem':
  print('A k√≥d nem fut le.')

else:
  print('Nem megfelel≈ë bemeneti √©rt√©k!')
````

## Sz√°m parit√°s√°nak vizsg√°lata

````python
szam = int(input('K√©rlek add meg a vizsg√°land√≥ sz√°mot: '))

print(f'A sz√°m marad√©ka 2-vel osztva: {szam % 2}')

if szam % 2 == 0:
  print('A megadott sz√°m p√°ros.')
else:
  print('A megadott sz√°m p√°ratlan.')
````

## Teszt √©rdemjegy ki√≠r√°sa pont alapj√°n

Az al√°bbi ponthat√°rokkal fogunk dolgozni:

0 - 40: el√©gtelen  
41 - 55: el√©gs√©ges  
56 - 70: k√∂zepes  
71 - 85: j√≥  
86 - 100: jeles

````python
pontszam = int(input("A teszten megszerzett pontsz√°mod: "))

if pontszam > 85:
  print('Jeles')
elif pontszam > 70:
  print('J√≥')
elif pontszam > 55:
  print('K√∂zepes')
elif pontszam > 40:
  print('El√©gs√©ges')
else:
  print('El√©gtelen')
````

## Egym√°sba √°gyazott el√°gaz√°sok

````python
hus = False
zoldseg = False

if hus:
  if zoldseg:
    print('Szereted a h√∫st √©s a z√∂lds√©geket is.')
  else:
    print('Csak a h√∫st szereted, a z√∂ls√©geket nem.')
else:
  if zoldseg:
    print('Nem szereted a h√∫st, de szereted a z√∂lds√©geket.')
  else:
    print('A h√∫st √©s a z√∂ls√©geket sem szereted.')
````

````python
hus = False
zoldseg = False

if hus and zoldseg:
  print('Szereted a h√∫st √©s a z√∂lds√©geket is.')
elif hus and not zoldseg:
  print('Csak a h√∫st szereted, a z√∂ls√©geket nem.')
elif not hus and zoldseg:
  print('Nem szereted a h√∫st, de szereted a z√∂lds√©geket.')
else:
  print('A h√∫st √©s a z√∂ls√©geket sem szereted.')
````

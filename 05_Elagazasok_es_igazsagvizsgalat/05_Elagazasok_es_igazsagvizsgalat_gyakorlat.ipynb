# Elágazás (if)

```python
szam_1 = float(input('Osztandó szám: '))
szam_2 = float(input('Osztó szám: '))

if szam_1 % szam_2 == 0:
  print(f'{szam_1}-ben megvan egésszer {szam_2}')
else:
  print(f'{szam_1}-ben nincs meg egésszer {szam_2}')
```

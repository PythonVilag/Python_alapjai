> A megadott hétköznap nem lézetik.


# 1. Feladat megoldás

```python
# Adat bekérése és megfelelő típusra alakítás
szam = int(input('Kérlek add meg a számot: '))

# Oszthatóság vizsgálata és eredmény kiírása
if szam % 3 == 0:
  print(f'A szám {szam} osztható 3-mal')
else:
  print(f'A szám {szam} nem osztható 3-mal')
```

# 2. Feladat megoldás

```python
# Adatok bekérése és megfelelő típusra alakítás
szam_1 = int(input('Kérlek add meg az egyik számot: '))
szam_2 = int(input('Kérlek add meg a másik számot: '))

# Oszthatóság vizsgálata és eredmény kiírása
if szam_1 % szam_2 == 0:
  print(f'A {szam_2} oszója a {szam_1}-nek')
else:
  print(f'A {szam_2} nem oszója a {szam_1}-nek')
```

# 3. Feladat megoldás

```python
# Adat bekérése és megfelelő típusra alakítás
szam = int(input('Kérlek add meg a számot: '))

# Oszthatóság vizsgálata és eredmény kiírása
if (szam % 3 == 0) and (szam % 5 == 0):
  print('FizzBuzz')
elif szam % 3 == 0:
  print('Fizz')
elif szam % 5 == 0:
  print('Buzz')
else:
  print(szam)
```

# 4. Feladat megoldás

```python
# Adatok bekérése és megfelelő típusra alakítás
szam_1 = float(input('Kérlek add meg az egyik számot: '))
szam_2 = float(input('Kérlek add meg a másik számot: '))
muvelet = input('Kérlek add meg a kívánt műveletet (+, -, *, /): ') # Nem kell átalakítani, mert jó a string formátum
eredmeny = False

# Művelet vizsgálata
if muvelet == '+':
  eredmeny = szam_1 + szam_2
elif muvelet == '-':
  eredmeny = szam_1 - szam_2
elif muvelet == '*':
  eredmeny = szam_1 * szam_2
elif muvelet == '/':
  eredmeny = szam_1 / szam_2

# Eredmény kiírása
if eredmeny:
  print(f'{szam_1} {muvelet} {szam_2} = {eredmeny:.3f}')
else:
  print('Valami adat rosszul lett megadva, kérlek javítsd')
```

# 5. Feladat megoldás

```python
# Adat bekérése és megfelelő típusra alakítás
nap = input('Kérlek adj meg egy hétkönapot: ')

napok = ['hétfő', 'kedd', 'szerda', 'csütörtök', 
         'péntek', 'szombat', 'vasárnap']

nap = nap.lower() # Átalakítjuk kisbetűssé, hogy ne számítson a kis és nagybetű

if nap in napok:
  print('A megadott hétköznap lézetik.')
else:
  print('A megadott hétköznap nem lézetik.')
```

# 1. Feladat

- Olvass be 2 (tört) számot és egy műveleti jelet (+, -, *, /)
- Számold ki 2 két számon végrehajtott művelet eredményét
- Írd ki az eredményt f-stringben 3 tízedes pontossággal

Példa:
> Kérlek add meg az egyik számot: 20.333  
> Kérlek add meg a másik számot: 4  
> Kérlek add meg a kívánt műveletet (+, -, *, /): /  
> 20.333 / 4.0 = 5.083

# 1. Megoldás

```python
# Adatok bekérése és megfelelő típusra alakítás
szam_1 = float(input('Kérlek add meg az egyik számot: '))
szam_2 = float(input('Kérlek add meg a másik számot: '))
muvelet = input('Kérlek add meg a kívánt műveletet (+, -, *, /): ') # Nem kell átalakítani, mert jó a string formátum
eredmeny = None

# Művelet vizsgálata
if muvelet == '+':
  eredmeny = szam_1 + szam_2
elif muvelet == '-':
  eredmeny = szam_1 - szam_2
elif muvelet == '*':
  eredmeny = szam_1 * szam_2
elif muvelet == '/':
  eredmeny = szam_1 / szam_2

# Eredmény kiírása
if eredmeny:
  print(f'{szam_1} {muvelet} {szam_2} = {eredmeny:.3f}')
else:
  print('Valami adat rosszul lett megadva, kérlek javítsd')
```

# 2. Feladat

- Kérj be egy egész számot
- Nézd meg, hogy osztható-e 2-vel
- Írasd ki a számot, és hogy páros vagy páratlan-e

Példa:
> Kérlek add meg a számot: 11  
> A szám 11 páratlan

# 2. Megoldás

```python
# Adat bekérése és megfelelő típusra alakítás
szam = int(input('Kérlek add meg a számot: '))

# Paritás vizsgálata és eredmény kiírása
if szam % 2 == 0:
  print(f'A szám {szam} páros')
else:
  print(f'A szám {szam} páratlan')
```

# 3. Feladat

- Kérj be két egész számot
- Vizsgáld meg, hogy a másodjára bekért szám osztója-e az elsőnek
- Írasd ki, hogy osztója, vagy sem

Példa:
> Kérlek add meg az egyik számot: 42  
> Kérlek add meg a másik számot: 7  
> A 7 oszója a 42-nek

# 3. Megoldás

```python
# Adatok bekérése és megfelelő típusra alakítás
szam_1 = int(input('Kérlek add meg az egyik számot: '))
szam_2 = int(input('Kérlek add meg a másik számot: '))

# Oszthatóság vizsgálata és eredmény kiírása
if szam_1 % szam_2 == 0:
  print(f'A {szam_2} oszója a {szam_1}-nek')
else:
  print(f'A {szam_2} nem oszója a {szam_1}-nek')
```

# 4. Feladat

- Kérj be egy egész számot
- Ha a szám osztható 3-al írd ki a kijelzőre, hogy 'Fizz'
- Ha a szám osztható 5-el írd ki, hogy 'Buzz'
- Ha szám 3-al és 5-el is osztható írd ki, hogy 'FizzBuzz'
- Minden más esetben írd ki a bemeneti számot

Példa:
> Kérlek add meg a számot: 30  
> FizzBuzz

# 4. Megoldás

```python
# Adat bekérése és megfelelő típusra alakítás
szam = int(input('Kérlek add meg a számot: '))

# Oszthatóság vizsgálata és eredmény kiírása
if (szam % 3 == 0) and (szam % 5 == 0):
  print('FizzBuzz')
elif szam % 3 == 0:
  print('Fizz')
elif szam % 5 == 0:
  print('Buzz')
```

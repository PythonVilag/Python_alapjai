# Lista

A lista egy olyan változó típus, ami tetszőleges elemek tárolására alkalmas.

Tulajdonságai:
- Az elemek sorrendben helyezkednek el
- Az elemekre a sorszámuk segítségével lehet hivatlkozni kapcsos zárójelek között: *\[ \]*
- A lista elemeit meg tudjuk változtatni
- Tetszőleges típusú elemeket képes tárolni (lista eleme is lehet egy listának)
- A lista számozása 0-val kezdődik!
- Az utólsó elemre -1. elemként tudunk hivatkozni (utólsó előtte -2 ...)

## Deklarálás (kapcsols zárójellel)

```python
lista_1 = [11, 'alma', 42.42]
lista_2 = [False, ['banán', 3], 'December']
```

### Új elem hozzáfűzése a listához

Új elemet vagy a *+* operátorral tudunk a listához adni, vagy az *append()* függvénnyel. 

- Fontos, hogy összeadás esetén az új elemnek is lista típusúnak kell lennie, illetve az úgy kapott listát el kell menteni egy új változóba

- append() esetén bármilyen típust hozzáfűzhetünk a listához, és a lista automatikusan módosul, nem kell új változóba elmenteni (in-place)

Összeadással

```python
# lista + 'új_elem'
```

```python
lista + ['új_elem']
```

```python
lista
```

<sub>A listának nem része 'új elem', mert nem mentettük el új változóba</sub>

Appendálás

```python
lista.append('új_elem')
```

<sub>Nincs visszatérési értéke a cellának, de a lista változó módosult</sub>

```python
lista
```

### Elem vizsgálata, hogy jelen van e a listában

### Egyéb

Néhány a népszerűbb listákon futtatható függvények közül:

- insert(index, elem) → beszúr egy elemet a megadott index helyére

- remove(elem) → eltávolítja az elemet a listából

- count(keresett_elem) → megadja, hogy keresett elem hányszor van jelen a listában

- reverse() → megfordítja a listában lévő elemek sorrendjét

```python
lista.insert(2, 'új kettes indexű elem')

print(lista)
```

```python
lista.remove('új_elem')

print(lista)
```

```python
lista.count('November')
```

```python
lista.reverse()

print(lista)
```

Csak azonos típusú listákon elvégezhető függvények:
- min()/max() → megadja a listában szereplő elemek közűl a minimálisat/maximálisat

- sort() → sorbarendezi a listát az elemek értéke alapján

- sum() → visszadja a lista elemeinek az összegét

```python
homogen_lista = [2.1, 5, 1, 6.5, 3]
```

```python
min(homogen_lista)
```

```python
max(homogen_lista)
```

```python
homogen_lista.sort()

print(homogen_lista)
```

```python
sum(homogen_lista)
```

# Lista

A lista egy olyan változó típus, ami elemek tárolására alkalmas.

Tulajdonságai:
- Az elemek sorrendben helyezkednek el
- Az elemekre a sorszámuk segítségével lehet hivatlkozni kapcsos zárójelek között: *\[ \]*
- A lista elemeit meg tudjuk változtatni
- Tetszőleges típusú elemeket képes tárolni (lista eleme is lehet egy listának)
- A lista számozása 0-val kezdődik!
- Az utólsó elemre -1. elemként tudunk hivatkozni (utólsó előtte -2 ...)

### Fontosabb függvények

Néhány a népszerűbb listákon futtatható függvények közül:

- ```<lista>.insert(<index>, <elem>)``` → beszúr egy elemet a megadott index helyére

- ```<lista>.remove(<elem>)``` → eltávolítja az elemet a listából

- ```<lista>.count(<keresett elem>)``` → megadja, hogy keresett elem hányszor van jelen a listában

- ```<lista>.reverse()``` → megfordítja a listában lévő elemek sorrendjét

````python
lista.insert(2, 'új kettes indexű elem')

print(lista)
````

````python
lista.remove('új_elem')

print(lista)
````

````python
lista.count('December')
````

````python
lista.reverse()

print(lista)
````

Csak azonos típusú elemekből alló listákon elvégezhető függvények:
- ```min(<lista>)```/```max(<lista>)``` → megadja a listában szereplő elemek közűl a minimálisat/maximálisat

- ```sort(<lista>)``` → sorbarendezi a listát az elemek értéke alapján

- ```sum(<lista>)``` → visszadja a lista elemeinek az összegét

````python
uniform_lista = [2.1, 5, 1, 6.5, 3]
````

````python
min(uniform_lista)
````

````python
max(uniform_lista)
````

````python
uniform_lista.sort()

print(uniform_lista)
````

````python
sum(uniform_lista)
````

# Set

A set lényegében egy olyan dictionary, amiben csak kulcsok vannak értékek nélkül.

Ez ezért tud hasznos lenni, mert már a dictionary-nál is, egy kulcs csak 1x szerepelhetett. Vagyis ha egy listából el szeretnánk távolítani a duplikált elemeket annak a legyegyszerűbb módja, hogy átalakítjuk set-é vagy vissza listává.

<sup>Fontos, hogy bár a duplikátumok eltűnnek, de a listában lévő elemek sorrendje is megcserélődik!</sup>

## Deklarálás

````python
set_pelda = {'alma', 'alma', 'barack', 'banán'}
````

````python
set_pelda
````

## Műveletek

````python
duplicated_list = ['Süss', 'fel', 'Nap', 'fényes', 'Nap']
````

````python
non_duplicated_list = list(set(duplicated_list))

non_duplicated_list
````

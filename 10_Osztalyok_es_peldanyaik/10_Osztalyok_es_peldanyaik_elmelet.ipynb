# Osztály

- Lehetőségünk van a korábban tanult adattípusokat (`int`, `string` ...) kiegészíteni új, egyedi típusokkal.

- Az osztály tartalmazza a típusunkhoz tartozó összekapcsolódó változókat és függvényeket

## Létrehozás és tulajdonságok

- A

- B

```python
class <Osztály neve>:
  <osztályt leíró kód>
```

<sup>A változó- és függvénynevekkel ellentétben az osztályok esetén bevett, hogy az első betű nagybetű</sup>

# Példák

## Egy cég munkatársait leíró adattípus

````python
class Alkalmazott:
  def __init__(self, nev, beosztas, fizetes, szabadsagon_van):
    self.nev = nev
    self.beosztas = beosztas
    self.fizetes = fizetes
    self.szabadsagon_van = szabadsagon_van

  def fizeztes_emeles(self):
    self.fizetes = self.fizetes * 1.05
````

````python
alkalmazott_1 = Alkalmazott('Peti', 'takarító', 80, False)
alkalmazott_2 = Alkalmazott('Laci', 'igazgató', 120, False)
````

````python
print(alkalmazott_1.beosztas)
print(alkalmazott_2.szabadsagon_van)
````

````python
alkalmazott_1.fizeztes_emeles()
print(alkalmazott_1.fizetes)
````

# Örökés

````python
class MávAlkalmazott(Alkalmazott):
  def __init__(self, nev, beosztas, fizetes, szabadsagon_van, keses):
    super().__init__(nev, beosztas, fizetes, szabadsagon_van)
    self.keses = keses
````

````python
mav_alkalmazott_1 = MávAlkalmazott('Pista', 'masinizta', 90, False, 30)

print(mav_alkalmazott_1.keses)
````

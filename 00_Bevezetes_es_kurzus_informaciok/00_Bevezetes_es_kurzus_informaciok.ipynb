## Példa Python kódra

````python
# Külső forrású kód importálása
from datetime import datetime, time, timedelta
````

````python
# Jelenlegi idő kiszámítása
current_GMT = datetime.now()
hour_shift = timedelta(hours=2)
current_time = current_GMT + hour_shift

# Holnapig hátralevő idő kiszámítása
tomorrow = current_time + timedelta(days=1)
time_until_tomorrow = datetime.combine(tomorrow, time.min) - current_time

# Eredmény kiírása
print(f'Még {str(time_until_tomorrow).split(".")[0]} van hátra holnapig!')
````

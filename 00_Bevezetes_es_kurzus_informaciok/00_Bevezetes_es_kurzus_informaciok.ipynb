````python
# Ez egy minta kód cella futtatáshoz
from datetime import datetime, time, timedelta

check_date = datetime.now()
hour_shift = timedelta(hours=2)
check_date = check_date + hour_shift
check_time = check_date.time()

if  time(7, 0) < check_time < time(23, 0):
  print(f'Még csak {check_time.strftime("%H:%M:%S")} van. Most még lehet dolgozni!')
else:
  print(f'Már {check_time.strftime("%H:%M:%S")} van. Ideje aludni!')
````

A fenti kódot futtatva feltűnhet, hogy mindig az aktuális időt írja ki a program. Egy cellát akárhányszor futtathatunk, viszont fontos észben tartani, hogy nem a cellák egymáshoz képesti elhelyezkedése (alatta/felette), hanem azok futtatási sorrendje számít!

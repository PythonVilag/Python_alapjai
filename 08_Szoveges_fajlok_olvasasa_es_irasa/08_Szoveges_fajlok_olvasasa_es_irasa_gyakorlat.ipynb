# 1. Feladat

- Kérj be a felhasználótól tetszőleges számú egész számot spacekkel elválasztva
- Hozz létre egy `csillagok.txt` nevű szöveges dokumentumot, melynek annyi sora van, ahány egész számot adott meg a felhasználó és minden sorában a megadott számnak megfelelőször szerepel a csillag (*) karakter

Példa:
> Egész számok space-el elválasztva: 2 4 1 0 6 3   

`csillagok.txt`

```
**  
****  
*  
  
******
***

```



# 2. Feladat

- Olvasd be a `szamok_es_betuk.txt` fájl tartalmát
- Csak minden 3-al osztható  indexű sorban számok vannak, a többiben betűk vannak
- Írd ki a kijelzőre a beolvasott számok átlagát

Példa:  
`szamok_es_betuk.txt`

```
5  
a  
b  
1 
c
d
3
```
> Beolvasott számok átlaga: 3.0

# 3. Feladat

- Olvasd be a `koordinatak.txt` fájl tartalmát
- Minden egyes pontpárra (sorra) számold ki az origótól (0, 0) vett távolságát (Pitagoras), két tízedes pontossággal
- Írd ki a `koordinatak_pitagoras.txt` fájlba az a koordináta pontokat és a számolt távolságot space-el elválasztva

Példa:  
`koordinatak.txt`

```
4 6
6 5
7 2
13 9
0 4
```


`koordinatak_pitagoras.txt`

```
4 6 7.21
6 5 7.81
7 2 7.28
13 9 15.81
0 4 4.0
```

#  1. Feladat megoldás

````python
bemenet = input("Egész számok space-el elválasztva: ")

bemenet = bemenet.split()
bemenet = [int(elem) for elem in bemenet]

with open("csillagok.txt", "w") as kimeneti_file:
  for elem in bemenet:
    print("*" * elem, file=kimeneti_file)
````

# 2. Feladat megoldás

````python
szamok = []

with open("szamok_es_betuk.txt", "r") as forras_file:
  for index, elem in enumerate(forras_file):
    if index % 3 == 0:
      szamok.append(int(elem))

print(f"Beolvasott számok átlaga: {sum(szamok) / len(szamok)}")
````

# 3. Feladat megoldás

````python
with open("koordinatak.txt", "r") as forras_file:
  with open("koordinatak_pitagoras.txt", "w") as kimeneti_file:
    for sor in forras_file:
      sor = sor.strip().split()
      x = int(sor[0])
      y = int(sor[1])
      tavolsag = (x**2 + y**2)**(1/2)
      tavolsag = round(tavolsag, 2)
      
      kimeneti_sor = f"{x} {y} {tavolsag}"
      print(kimeneti_sor, file=kimeneti_file)
````

# File kezelés

Általános lépések:

- File megnyitása, a tartalom eléréséhez
- File-ból az adatok kiolvasás/file-ba az adatok írása
- File bezárása, hogy a a többi program is hozzáférjen

Struktúra:
```python
with open('file_neve', 'megnyitas_tipusa') as valtozo_neve:
  csinalj valamit valtozo_neve-vel
```

A `with` parancs automatikusan bezárja a file, és kezeli a futtatás közbeni esetleges hibákat

A megnyitás típusa általában olvasás: 'r' vagy írás 'w'

# Adatok kiírása

````python
'al m a'.split()

````

````python
gyumolcsok = ['alma   ', 'banán     ', 'narancs  ']
````

````python
with open('sample_data/text.txt', 'w') as f:
  for elem_szam, gyumolcs in enumerate(gyumolcsok):
    f.write(f'{elem_szam + 1}. {gyumolcs}\n')
````

# Adatok beolvasása

````python
ures_lista = []
````

````python
with open('sample_data/text.txt', 'r') as f:
  for sor in f:
    sor = sor.strip()
    ertek = int(sor[0])
    ures_lista.append(ertek)
````

````python
ures_lista
````

````python
sum(ures_lista)
````

FizzBuzz

````python
for szam in range(1, 17):
  if (szam % 3 == 0) and (szam % 5 == 0):
    print('FizzBuzz')
  elif szam % 3 == 0:
    print('Fizz')
  elif szam % 5 == 0:
    print('Buzz')
  else:
    print(szam)
````

````python
for szam in range(1, 17):
  kimenet = ''
  if szam % 3 == 0:
    kimenet += 'Fizz'
  if szam % 5 == 0:
    kimenet += 'Buzz'

  if len(kimenet) == 0:
    kimenet = szam

  print(kimenet)

  
````

# 5. Feladat

- Kérj be egy zárójel sorozatot
- Írd ki, hogy a zárójelezés megfelelő-e, azaz mindig csak megnyitott zárójelet zárunk be és az összes megnyitott zárójel bezárásra kerül

Példa:
> Add meg a vizsgálandó szöveget: (())()  
> A képlet zárójelezése rendben van.

# 1. Feladat megoldás

```python
felso_hatar = int(input('Kérlek adj meg egy egész számot: '))

for szam in range(1, felso_hatar + 1):
  if (szam % 3 == 0) and (szam % 5 == 0):
    print('FizzBuzz')
  elif (szam % 3 == 0):
    print('Fizz')
  elif (szam % 5 == 0):
    print('Buzz')
  else:
    print(szam)
```

# 2. Feladat megoldás

```python
osztando_szam = int(input('Kérlek adj meg egy egész számot: '))

print(f'{osztando_szam} osztói: ')

for szam in range(1, osztando_szam + 1):
  if osztando_szam % szam == 0:
    print(szam)
```

# 3. Feladat megoldás

```python
bekert_szam = int(input('Kérlek adj meg egy egész számot: '))
osztok_szama = 0

for szam in range(1, bekert_szam + 1):
  if bekert_szam % szam == 0:
    osztok_szama += 1


if osztok_szama == 2:
  print(f'A(z) {bekert_szam} pírm.')
else: 
  print(f'A(z) {bekert_szam} nem pírm.')
```

# 4. Feladat megoldás

```python
titkos_szam = 11

tipp = None

while tipp != titkos_szam:
  tipp = int(input('Tipp: '))
  if tipp < titkos_szam:
    print('Nagyobb!')
  if tipp > titkos_szam:
    print('Kisebb!')

print(f'Kitaláltad! A titkos szám a {titkos_szam}!')
```

# 5. Feladat megoldás

```python
vizsgalando_szoveg = input('Add meg a vizsgálandó szöveget: ')

elkezdett_zarojelek = 0
hibas_keplet = False

for betu in vizsgalando_szoveg:
  if betu == '(':
    elkezdett_zarojelek += 1
  if betu == ')':
    elkezdett_zarojelek -= 1

  if elkezdett_zarojelek < 0:
    hibas_keplet = True
    break

if elkezdett_zarojelek > 0:
  hibas_keplet = True

if hibas_keplet:
  print('Hibás képlet!')
else:
  print('A képlet zárójelezése rendben van.')
```

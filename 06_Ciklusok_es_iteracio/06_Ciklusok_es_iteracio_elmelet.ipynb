# Ciklus fajták és jellemzőik

- Parancsok többszöri végrehajtása esetén használjuk
- Addig hajtódik végre, amíg a feltétel igaz

Két fő típusa van:
- A végrehajtások számát ismerjük → for ciklus
- Nem tudjuk hányszor kell az adott kódot végrehajtani → while ciklus

<sub>Sokszor mindkét típusú problémát meg tudjuk oldani bármely ciklus használatával, viszont az olvashatóságot könnyíti a fentieknek megfelelő használat.</sub>

# While ciklus

- Akkor használjuk, ha a pontos végrehatási szám nem ismert
- Addig hajtódik végre, amíg a futás feltétele teljesül (szélsőséges esetek: 0 - $\infty$)
- A kódnak ki kell hatnia a feltételre (különben mindig végtelenszer fog futni)

Struktúra:
```python
while <feltétel>:
  <csinálj valamit amíg "feltétel" logikai érétke igaz>
  <módosítsd a feltételben szereplő kifejezést>
```

````python
szam = 0

while szam < 12:
  print(szam)
  szam += 1    # szam = szam + 1
````

# For ciklus

- A végrehajtás darabszáma ismert (pl. listán való végigiterálás)
- A ciklus 'magától léptet', nincs szükség feltételre és azt módosító kódrészletre

Struktúra:
```python
for <ideiglenes változó> in <lista neve>:
  <csinálj valamit az "ideiglenes változó" segítségével>
```

<sup>Nem csak listán tudunk végigiterálni, hanem az összes többi adat tároló típuson is (tuple, set, dictionary) és még stringeken is.</sup>

````python
gyumolcsok = ['alma', 'banán', 'narancs']

for gyumolcs in gyumolcsok:
  print(gyumolcs)
````

Range

````python
for szam in range(3):
  print(szam)
````

Egymásba ágyazott ciklusok

````python
for szam in range(3):
  for gyumolcs in gyumolcsok:
    print(szam, gyumolcs)
````

Enumerate

````python
gyumolcsok = ['alma', 'banán', 'narancs']

for elem_szam, gyumolcs in enumerate(gyumolcsok):
  print(elem_szam, gyumolcs)
````

Lista létrehozása ciklussal

````python
szamok = [1, 3, 5]

szamok_negyzete = [szam**2 for szam in szamok]
print(szamok_negyzete)
````

````python
szamok_negyzete = []
for szam in szamok:
  szamok_negyzete.append(szam**2)
````

````python
szovegek = ['1', '2', '3']
szamok = []

for elem in szovegek:
  szamok.append(int(elem))
````

````python
szamok = [int(elem) for elem in szovegek]
````

````python
szamok
````

````python
szamok = list(map(int, szovegek))
````

````python
szamok
````

````python
def add_two(num):
  result = num + 2
  return result

szamok = list(map(add_two, szamok))
szamok
````

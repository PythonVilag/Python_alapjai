# Ciklus fajt√°k √©s jellemz≈ëik

- Parancsok t√∂bbsz√∂ri v√©grehajt√°sa eset√©n haszn√°ljuk

- K√©t f≈ë t√≠pusa van:

  - Nem tudjuk h√°nyszor kell az adott k√≥dot v√©grehajtani ‚Üí `while` ciklus

  - A v√©grehajt√°sok sz√°m√°t ismerj√ºk ‚Üí `for` ciklus

<sub>Sokszor mindk√©t t√≠pus√∫ probl√©m√°t meg tudjuk oldani b√°rmely ciklus haszn√°lat√°val, viszont a meg√©rt√©st k√∂nny√≠ti a fentieknek megfelel≈ë haszn√°lat.</sub>

# While ciklus

- Akkor haszn√°ljuk, ha a pontos v√©grehajt√°si sz√°m nem ismert

- Addig hajt√≥dik v√©gre, am√≠g a fut√°s felt√©tele teljes√ºl (sz√©ls≈ës√©ges esetek: 0 - $\infty$)

- A k√≥dnak ki kell hatnia a felt√©telre (k√ºl√∂nben mindig v√©gtelenszer fog futni)

## L√©trehoz√°s √©s tulajdons√°gok

- While ciklust ~~nem meglep≈ë m√≥don~~ a `while` kulcssz√≥ seg√≠ts√©g√©vel tudunk l√©trehozni

- Az el√°gaz√°shoz hasonl√≥an itt is egy felt√©tel k√∂veti a parancsot, majd kett≈ëspont √©s a bentebb h√∫zott k√≥dr√©szlet

- A bentebb h√∫zott k√≥dr√©szlet fontos, hogy kihasson a felt√©telre, hogy a ciklus id≈ëvel lefusson


```python
while <felt√©tel>:
  <csin√°lj valamit am√≠g <felt√©tel> logikai √©rt√©ke igaz>
  <m√≥dos√≠tsd a felt√©telben szerepl≈ë kifejez√©st>
```

<sup>A felt√©tel ut√°n kett≈ëspont van √©s az adott felt√©telhez tartoz√≥ k√≥d (az el√°gaz√°shoz hasonl√≥an) bentebb van h√∫zva!</sup>

<sup>Ez jelenti azt, hogy a melyik sor tartozik m√©g a ciklushoz √©s melyik m√°r nem.</sup>

````python
megfelelo_bemenet = False

while not megfelelo_bemenet:
  valasztas = input('Szereted a csokik? (igen/nem)')

  if (valasztas == 'igen') or (valasztas == 'nem'):
    megfelelo_bemenet = True
  else:
    print('Nem megfelel≈ë bemenet!')


if valasztas == 'igen':
  print('\nTe szereted a csokit!')
else:
  print('\nTe nem szereted a csokit!')
````

# For ciklus

- A v√©grehajt√°s darabsz√°ma ismert (pl. stringen/list√°n val√≥ v√©gig iter√°l√°s)

- A ciklus 'mag√°t√≥l l√©ptet', nincs sz√ºks√©g felt√©telre √©s azt m√≥dos√≠t√≥ k√≥d r√©szletre

## L√©trehoz√°s √©s tulajdons√°gok

- For ciklust ~~nem meglep≈ë m√≥don~~ a `for` kulcssz√≥ seg√≠ts√©g√©vel tudunk l√©trehozni

- Ezt k√∂veti egy ideiglenes v√°ltoz√≥ neve amivel a l√©ptet√©skor felvett k√ºl√∂nb√∂z≈ë √©rt√©kekre fogunk tudni hivatkozni

- Ezut√°n a `in` kulcssz√≥ k√∂vetkezik, majd annak a list√°nak (vagy m√°sik t√≠pusnak) a neve, amin v√©gig szeretn√©nk iter√°lni.

- Az el√°gaz√°shoz √©s a while ciklushoz hasonl√≥an a sor v√©ge itt is egy kett≈ëspont amit a bentebb h√∫zott k√≥dr√©szlet k√∂vet


```python
for <ideiglenes v√°ltoz√≥> in <lista neve>:
  <csin√°lj valamit az <ideiglenes v√°ltoz√≥> seg√≠ts√©g√©vel>
```

<sup>A felt√©tel ut√°n kett≈ëspont van √©s az adott felt√©telhez tartoz√≥ k√≥d (az el√°gaz√°shoz hasonl√≥an) bentebb van h√∫zva!</sup>

<sup>Ez jelenti azt, hogy a melyik sor tartozik m√©g a ciklushoz √©s melyik m√°r nem.</sup>

<sup>Nem csak stringen √©s list√°n tudunk v√©gig iter√°lni, hanem az √∂sszes t√∂bbi adat t√°rol√≥ t√≠puson is (`list`, `dictionary`, ...).</sup>

## P√©ld√°k

### List√°n val√≥ v√©gig iter√°l√°s

Az ideiglenes v√°ltoz√≥nk √©rt√©ke felveszi a lista √∂sszes elem√©nek az √©rt√©k√©t.

````python
gyumolcsok = ['alma', 'ban√°n', 'narancs']

for gyumolcs in gyumolcsok:
  print(gyumolcs)

print(gyumolcs)
````

### Sz√°m intervallumon val√≥ v√©gig l√©peget√©s (`range`)

A `range()` parancs seg√≠ts√©g√©vel lehet≈ës√©g√ºnk van sz√°mok k√∂z√∂tt iter√°lni sok tov√°bbi be√°ll√≠t√°ssal.

````python
for szam in range(3):
  print(szam)
````

````python
for szam in range(3, 9):
  print(szam)
````

<sup>A list√°k √©s stringket indexel√©s√©hez hasonl√≥an itt is alulr√≥l inkluz√≠v, fel√ºlr≈ël pedig exkluz√≠v az intervallum.</sup>

````python
for szam in range(3, 9, 2):
  print(szam)
````

````python
for szam in range(len(gyumolcsok)):
  print(gyumolcsok[szam])
````

````python
alap = 3
kitevo = 4

eredmeny = 1
for _ in range(kitevo):
  eredmeny *= alap

print(eredmeny)
````

### Iter√°l√°s az index lek√©r√©se mellett (`enumerate`)

Az `enumerate()` egy nagyon hasznos f√ºggv√©ny, aminek a seg√≠ts√©g√©vel iter√°l√°s k√∂zben azt is nyomon tudjuk k√∂vetni, hogy h√°nyadik elemn√©l j√°runk √©ppen.

````python
gyumolcsok = ['alma', 'ban√°n', 'narancs']

for elem_szam, gyumolcs in enumerate(gyumolcsok):
  print(elem_szam, gyumolcs)
````

````python
gyumolcsok = ['alma', 'ban√°n', 'narancs']

for elem_szam, gyumolcs in enumerate(gyumolcsok):
  if elem_szam == 1:
    print(f'A kedvenc gy√ºm√∂lcs√∂m a {gyumolcs}.')
````

### Lista l√©trehoz√°sa ciklussal (üêü)

√öj lista l√©trehoz√°sa, egy kor√°bbi seg√≠ts√©g√©vel ahol az lista elemeket n√©gyzetre emelj√ºk.

````python
szamok = [1, 3, 5]
szamok_negyzete = []

for szam in szamok:
  szamok_negyzete.append(szam**2)

print(szamok_negyzete)
````

````python
szamok = [1, 3, 5]
szamok_negyzete = [szam**2 for szam in szamok]

print(szamok_negyzete)
````

√öj lista l√©trehoz√°sa, egy kor√°bbi seg√≠ts√©g√©vel ahol az lista elemeket eg√©sz sz√°mm√° alak√≠tjuk.

````python
szovegek = ['1', '2', '3']
szamok = []

for elem in szovegek:
  szamok.append(int(elem))

print(szamok)
````

````python
szovegek = ['1', '2', '3']

szamok = [int(elem) for elem in szovegek]

print(szamok)
````

Map (üêü)

A `map` parancsnak √°t tudunk adni egy f√ºggv√©nyt √©s egy iter√°lhat√≥ t√≠pust (jelen esetben list√°t) √©s a kimenetet list√°v√° alak√≠tva megkapjuk az eredeti lista minden elem√©n lefuttatot f√ºggv√©ny kimenet√©t.

````python
szovegek = ['1', '2', '3']

szamok = list(map(int, szovegek))

print(szamok)
````

# Break √©s continue (üêü)

Bizonyos felt√©telek teljes√ºl√©se eset√©n lehet≈ës√©g√ºnk van az eg√©sz ciklusb√≥l kil√©pni, vagy a k√∂vetkez≈ë iter√°ci√≥ra ugrani.

## P√©ld√°k

### Kil√©p√©s for ciklusb√≥l bizonyos elem eset√©n

````python
lista = [1, 2, 'alma', False, 5, True]

for elem in lista:
  if elem == False:
    print('Hamis √©rt√©k megtal√°lva!')
    break
  print(elem)

print(elem)
````

### Kil√©p√©s while ciklusb√≥l a felt√©tel teljes√≠t√©se n√©lk√ºl

````python
szam = 1

while True:
  szam *= 2
  print(szam)
  if szam > 600:
    break
````
